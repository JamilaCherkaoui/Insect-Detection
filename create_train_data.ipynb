```python
import os
from PIL import Image

# Dossier contenant les fichiers .txt et .jpg
BASE_DIR = '/Users/alyazouzou/Desktop/ChallengeIA/data'  # ← à adapter
OUTPUT_DIR = os.path.join('/Users/alyazouzou/Desktop/ChallengeIA/', 'imagettes')
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Lister tous les fichiers .txt
txt_files = [f for f in os.listdir(BASE_DIR) if f.endswith('.txt')]

for txt_file in txt_files:
    labels = os.path.splitext(txt_file)[0]  # Ex: "4_4_4_4"
    image_file = f"{labels}.jpg"
    image_path = os.path.join(BASE_DIR, image_file)
    txt_path = os.path.join(BASE_DIR, txt_file)

    if not os.path.exists(image_path):
        print(f"🚫 Image non trouvée : {image_file}")
        continue

    with open(txt_path, 'r') as f:
        lines = f.readlines()

    for line in lines:
        parts = line.strip().split()
        if len(parts) < 6:
            print(f"❌ Ligne ignorée : {line}")
            continue

        # Lire les coordonnées (les 4 derniers)
        x_center, y_center, width, height = map(float, parts[-4:])
        projet_annee = " ".join(parts[:-4]).replace('/', '_')

        # Ouvrir image
        img = Image.open(image_path)
        W, H = img.size

        # YOLO → pixel
        x = x_center * W
        y = y_center * H
        w = width * W
        h = height * H

        left = max(0, int(x - w / 2))
        top = max(0, int(y - h / 2))
        right = min(W, int(x + w / 2))
        bottom = min(H, int(y + h / 2))

        crop = img.crop((left, top, right, bottom))

        # Nom du fichier imagette
        output_name = f"{labels}-{projet_annee}-{x_center}-{y_center}-{width}-{height}.jpg"
        output_path = os.path.join(OUTPUT_DIR, output_name)
        crop.save(output_path)
        print(f"✅ Sauvé : {output_name}")
```
